```python
A={"a":0.2,"b":0.3,"c":0.4,"d":0.5}
B={"a":0.7,"b":0.5,"c":0.9,"d":0.8}
u={}
x={}
print("Fuzzy set A : ",A)
print("Fuzzy set B : ",B)

print("\nLHS : ")
def union(A,B):
    for i in A and B:
        x[i]=max(A[i],B[i])
    return x
print("\n(AUB): ",union(A,B))

def compliment(x):
    for i in x:
        u[i]=round(1-x[i],2)
    return u
a=compliment(x)
print("\n(AUB)': ",a)

print("\nRHS : ")

def compliment(A):
    for i in A:
        u[i]=round(1-A[i],2)
    return u
y=compliment(A)
print("\nA' : ",y)
z=compliment(B)
print("\nB' : ",z)

def intersection(x,y):
    for i in x and y:
        u[i]=min(x[i],y[i])
    return u
b=intersection(x,y)
print("\nA' ∩ B': ",b)

if a==b:
    print("\n(AUB)'= A' ∩ B' :",a,"=",b)
    print("\nDe Morgan's Law has Proved")
else:
    print("De Morgan's Law has not Proved")
```
